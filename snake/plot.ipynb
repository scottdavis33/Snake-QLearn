```python
import matplotlib.pyplot as plt
import numpy as np

from agent import train  # Import the train function from your Python script

scores, mean_scores, records = train(max_games = 5000)  # Run your training function

print(max(mean_scores))

# Now, plot the results using the same plotting code you have
plt.figure(figsize=(10, 5))
plt.title('Record, Avg. Score, and Scores vs. # of Games')
plt.xlabel('Number of Games')
plt.ylabel('Score')
plt.plot(scores, color='blue', label='Scores')
plt.plot(mean_scores, color='green', label='Mean Scores')
plt.plot(records, color='red', label='Records')

# Adding text annotations for the last values on the plot
plt.text(len(scores)-1, scores[-1], str(scores[-1]), color='blue')
plt.text(len(mean_scores)-1, mean_scores[-1], str(mean_scores[-1]), color='green')
plt.text(len(records)-1, records[-1], str(records[-1]), color='red')

plt.legend()
plt.show()
```
